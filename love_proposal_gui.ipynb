```python
import tkinter as tk
from tkinter import messagebox

def say_yes():
    messagebox.showinfo(
        "üíç She Said YES üíç",
        "You just made me the happiest person ‚ù§Ô∏è\nI love you forever ü•∞"
    )

def say_no():
    messagebox.showwarning(
        "üíî",
        "Error 404: Love Not Found üò¢\nBut my love will never stop ‚ù§Ô∏è"
    )

root = tk.Tk()
root.title("üíñ My Python Proposal üíñ")
root.geometry("400x300")
root.configure(bg="pink")

label = tk.Label(
    root,
    text="Hey My Love ‚ù§Ô∏è\nWill you be my girlfriend? üíç",
    font=("Arial", 16, "bold"),
    bg="pink",
    fg="red"
)
label.pack(pady=30)

yes_btn = tk.Button(
    root,
    text="YES ‚ù§Ô∏è",
    font=("Arial", 14, "bold"),
    bg="red",
    fg="white",
    command=say_yes
)
yes_btn.pack(side="left", padx=40, pady=20)

no_btn = tk.Button(
    root,
    text="NO üíî",
    font=("Arial", 14, "bold"),
    bg="white",
    fg="black",
    command=say_no
)
no_btn.pack(side="right", padx=40, pady=20)

root.mainloop()
```
